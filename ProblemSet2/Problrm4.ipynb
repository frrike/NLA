SVD decomposition of Lena and it's singular values

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc

lena = misc.lena()


u, s, v = np.linalg.svd(lena)
fig, ax = plt.subplots(1, 1)
ax.semilogy(s[:])
```

```python
ranks = [10, 20, 50, 100, 300, 512]

titles = ['{} singular values'.format(r) for r in ranks]
lena_appr = [u[:,:r].dot(np.diagflat(s[:r])).dot(v[:r,:]) for r in ranks]
    
fig, ax = plt.subplots(3, 2, figsize=(20, 20))
for i in range(len(ranks)):
    ax[(i) / 2, (i) % 2].imshow(lena_appr[i], cmap='gray')
    ax[(i) / 2, (i) % 2].set_title(titles[i])
fig.tight_layout()
```
